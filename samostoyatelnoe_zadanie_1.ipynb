```python
# Анализатор текста
# Запрашиваем текст, пока длина не будет минимум 20 символов
while True:
    text = input("Введите текст (минимум 20 символов): ")
    if len(text) >= 20:
        break
    else:
        print(f"Слишком коротко! Введите ещё {20 - len(text)} символов.\n")

# 1. Длина текста
print(f"\nДлина текста: {len(text)} символов")

# 2. Текст в верхнем и нижнем регистре
print(f"В верхнем регистре: {text.upper()}")
print(f"В нижнем регистре: {text.lower()}")

# 3. Количество слов (разделитель - пробел)
words = text.split()
print(f"Количество слов: {len(words)}")

# 4. Первое и последнее слово 
if words:
    print(f"Первое слово: {words[0]}")
    print(f"Последнее слово: {words[-1]}")
else:
    print("Нет слов (только пробелы?)")

# 5. Проверка, содержит ли текст слово "Python" 
if "python" in text.lower():
    print("Текст содержит слово 'Python' ")
else:
    print("Текст НЕ содержит слово 'Python'")

# 6. Замена пробелов на дефисы
print(f"Текст с дефисами вместо пробелов: {text.replace(' ', '-')}")
```
