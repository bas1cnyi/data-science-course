```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = sns.load_dataset('titanic')
df['age'] = df['age'].fillna(df['age'].median())


plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='class', hue='class', palette='viridis', legend=False)
plt.title('Количество пассажиров по классам')
plt.xlabel('Класс')
plt.ylabel('Количество')
plt.show()


plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='age', hue='sex', kde=True, bins=30, alpha=0.6)
plt.title('Распределение возраста по полу')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='age', hue='survived', 
            palette={0: 'salmon', 1: 'lightgreen'})
plt.title('Возраст пассажиров по классам и выживаемости')
plt.xlabel('Класс')
plt.ylabel('Возраст')
plt.legend(title='Выжил', labels=['Нет', 'Да'])
plt.show()


numeric_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True)
plt.title('Корреляционная матрица')
plt.show()
```
